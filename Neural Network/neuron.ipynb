# coding Neuron and Layers.

## 1. coding our first neuron.

### a. Neuron with 3 inputs.

```python
inputs = [1,2,3]
wights = [0.2,0.8,-0.5]
bias = 2
output = inputs[0]*wights[0] + inputs[1]*wights[1] + inputs[2]*wights[2] + bias 
print(output)
```

### b. Neuron with 4 inputs.

```python
inputs = [1,2,3,4]
wights = [0.2,0.8,-0.5,1]
bias = 2
output = inputs[0]*wights[0] + inputs[1]*wights[1] + inputs[2]*wights[2] + inputs[3]*wights[3] + bias 
print(output)
```

## 2. Layers in Neurons

### Layers with 4 inputs and 3 Neurons.

```python
inputs = [1,2,3,2.5]
wights = [[0.2,0.8,-0.5,1],
          [0.5,0.1,-0.3,0.7],
          [0.3,0.7,0.2,0.8]]
bias = [2,3,0.5]
output = [inputs[0]*wights[0][0] + inputs[1]*   wights[0][1] + inputs[2]*wights[0][2] + inputs[3]*wights[0][3] + bias[0],
          inputs[0]*wights[1][0] + inputs[1]*   wights[1][1] + inputs[2]*wights[1][2] + inputs[3]*wights[1][3] + bias[1],
          inputs[0]*wights[2][0] + inputs[1]*   wights[2][1] + inputs[2]*wights[2][2] + inputs[3]*wights[2][3] + bias[2]]

print(output)
```

### using loops build layers

```python
inputs = [1,2,3,2.5]
weights = [
    [0.2,0.8,-0.5,1],
    [0.5,0.1,-0.3,0.7],
    [0.3,0.7,0.2,0.8]
]
biases = [2,3,0.5]
outputs = []
for neuron_weights, neuron_bias in zip(weights, biases):
    neuron_output = 0
    for input, weight in zip(inputs,neuron_weights):
        neuron_output += input * weight
    neuron_output += neuron_bias
    outputs.append(neuron_output)
print(outputs)
```
