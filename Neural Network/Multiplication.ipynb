### different type of multiplicaton

- Vector * Vector 
   * it is considered as neurons 
- Vector * Matrix
    * it is considered as layers
- Matrix * Matrix
    * it is considered as batch.


Using numpy build neurons .

```python
import numpy as np 
inputs = [1,2,3,4]
weights = [0.2,0.8,-0.5,1]
bias = 2
output = np.dot(inputs, weights) + bias
print(output)
```

Using Numpy to build layers.

```python
import numpy as np
inputs = [1,2,3,4]
weights = [[0.2,0.8,-0.5,1],
           [0.5,0.1,-0.3,0.7],
           [0.9,0.4,0.2,0.5]]
bias = [2,3,0.5]
output = np.dot(weights, inputs) + bias
print(output)
```

Using Numpy build batch of data

```python
inputs =[[1,2,3,4],
         [2,3,4,5],
         [3,4,5,6]]
weights = [[0.2,0.8,-0.5,1],
           [0.5,0.1,-0.3,0.7],  
           [0.9,0.4,0.2,0.5]]
bias = [2,3,0.5]
output = np.dot( np.array(inputs),np.array(weights).T) + np.array(bias)
print(output)
```
